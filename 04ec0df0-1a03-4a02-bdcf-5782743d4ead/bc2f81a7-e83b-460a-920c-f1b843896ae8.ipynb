##화씨 온도 변환기
*   화씨 온도를 입력 받는다.
*   섭시 온도로 바꿔서 아래 예시처럼 출력한다.
*   C = (F-32)*5/9




```python
F=int(input("화씨온도:"))
C=round((F-32)*5/9,2) #round(숫자,자릿수)
print("섭씨온도:",C)
```

##사각형의 넓이와 둘레 계산
* 길이를 입력하시오 : 30
* 너비를 입력하시오 : 50
* 넓이 : 30*50=1500
* 둘레 : 30 * 2 + 50 * 2 = 160

```python
a = int(input("길이를 입력하시오 : "))
b = int(input("너비를 입력하시오 : "))
print("넓이 =",a ,"*",b,"=",a*b)
print("둘레 =",a,"* 2","+",b,"* 2","=",a*2+b*2)
```

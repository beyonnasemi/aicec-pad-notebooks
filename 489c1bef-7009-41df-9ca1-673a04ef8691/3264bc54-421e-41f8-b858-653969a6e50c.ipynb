```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# ============================================================
# 1. 분석 조건 설정
# ============================================================
file_name = "NASA Cycle DATA.xlsx"
sheet_names = ["B05", "B06", "B07", "B18"]

# 전 cycle보다 조금이라도 증가하면 Recovery로 판정
RECOVERY_THRESHOLD_AH = 0.000

# 측정 노이즈를 제외하고 싶으면 5 mAh로 변경
# RECOVERY_THRESHOLD_AH = 0.005


# ============================================================
# 2. 결과 저장용 변수
# ============================================================
all_data = {}
event_data = {}
interval_data = {}
summary_list = []


# ============================================================
# 3. 셀별 Recovery 분석
# ============================================================
for sheet_name in sheet_names:

    print("\n" + "=" * 60)
    print(f"{sheet_name} 분석")
    print("=" * 60)

    # --------------------------------------------------------
    # 데이터 불러오기
    # --------------------------------------------------------
    df = pd.read_excel(
        file_name,
        sheet_name=sheet_name
    )

    # 열 이름 앞뒤 공백 제거
    df.columns = df.columns.str.strip()

    # 필요한 열이 존재하는지 확인
    required_columns = [
        "Cycle number",
        "Discharge_Capacity"
    ]

    for column in required_columns:
        if column not in df.columns:
            raise ValueError(
                f"{sheet_name} 시트에 '{column}' 열이 없습니다."
            )

    # Sample 열이 없으면 시트 이름으로 생성
    if "Sample" not in df.columns:
        df["Sample"] = sheet_name

    # 숫자형으로 변환
    df["Cycle number"] = pd.to_numeric(
        df["Cycle number"],
        errors="coerce"
    )

    df["Discharge_Capacity"] = pd.to_numeric(
        df["Discharge_Capacity"],
        errors="coerce"
    )

    # 결측 행 제거
    df = df.dropna(
        subset=[
            "Cycle number",
            "Discharge_Capacity"
        ]
    )

    # Cycle 순으로 정렬
    df = df.sort_values(
        "Cycle number"
    ).reset_index(drop=True)


    # ========================================================
    # 4. 전 cycle 대비 용량 변화 계산
    # ========================================================
    df["Previous_Cycle"] = (
        df["Cycle number"].shift(1)
    )

    df["Previous_Capacity_Ah"] = (
        df["Discharge_Capacity"].shift(1)
    )

    # 현재 용량 - 이전 용량
    df["Delta_Capacity_Ah"] = (
        df["Discharge_Capacity"]
        - df["Previous_Capacity_Ah"]
    )

    # Ah → mAh
    df["Delta_Capacity_mAh"] = (
        df["Delta_Capacity_Ah"] * 1000
    )

    # 전 cycle 대비 Recovery percentage
    df["Event_Recovery_Percent"] = (
        df["Delta_Capacity_Ah"]
        / df["Previous_Capacity_Ah"]
        * 100
    )


    # ========================================================
    # 5. Recovery event 판정
    # ========================================================
    df["Recovery_Event"] = (
        df["Delta_Capacity_Ah"]
        > RECOVERY_THRESHOLD_AH
    )

    # Recovery 발생 행만 선택
    recovery_events = df[
        df["Recovery_Event"] == True
    ].copy()

    recovery_events = recovery_events[
        [
            "Sample",
            "Previous_Cycle",
            "Cycle number",
            "Previous_Capacity_Ah",
            "Discharge_Capacity",
            "Delta_Capacity_Ah",
            "Delta_Capacity_mAh",
            "Event_Recovery_Percent"
        ]
    ]


    # ========================================================
    # 6. 연속된 Recovery event를 하나의 구간으로 묶기
    # ========================================================
    previous_event = (
        df["Recovery_Event"]
        .shift(1, fill_value=False)
    )

    # 이전 행은 Recovery가 아니고 현재 행이 Recovery이면
    # 새로운 Recovery 구간 시작
    df["Recovery_Start"] = (
        df["Recovery_Event"]
        & (~previous_event)
    )

    # Recovery 구간 번호 부여
    df["Recovery_Group"] = (
        df["Recovery_Start"].cumsum()
    )

    recovery_rows = df[
        df["Recovery_Event"] == True
    ].copy()


    # ========================================================
    # 7. Recovery 구간별 계산
    # ========================================================
    if len(recovery_rows) > 0:

        recovery_intervals = (
            recovery_rows
            .groupby("Recovery_Group")
            .agg(
                Start_Cycle=(
                    "Previous_Cycle",
                    "first"
                ),
                End_Cycle=(
                    "Cycle number",
                    "last"
                ),
                Recovery_Event_Count=(
                    "Cycle number",
                    "count"
                ),
                Start_Capacity_Ah=(
                    "Previous_Capacity_Ah",
                    "first"
                ),
                End_Capacity_Ah=(
                    "Discharge_Capacity",
                    "last"
                ),
                Sum_Recovery_Ah=(
                    "Delta_Capacity_Ah",
                    "sum"
                ),
                Sum_Recovery_mAh=(
                    "Delta_Capacity_mAh",
                    "sum"
                ),
                Mean_Event_Recovery_Percent=(
                    "Event_Recovery_Percent",
                    "mean"
                ),
                Maximum_Event_Recovery_Percent=(
                    "Event_Recovery_Percent",
                    "max"
                )
            )
            .reset_index()
        )

        # 구간 시작 대비 종료 용량의 순증가량
        recovery_intervals["Net_Interval_Recovery_Ah"] = (
            recovery_intervals["End_Capacity_Ah"]
            - recovery_intervals["Start_Capacity_Ah"]
        )

        recovery_intervals["Net_Interval_Recovery_mAh"] = (
            recovery_intervals["Net_Interval_Recovery_Ah"]
            * 1000
        )

        # 구간 전체 Recovery percentage
        recovery_intervals["Interval_Recovery_Percent"] = (
            recovery_intervals["Net_Interval_Recovery_Ah"]
            / recovery_intervals["Start_Capacity_Ah"]
            * 100
        )

        recovery_intervals["Start_Cycle"] = (
            recovery_intervals["Start_Cycle"]
            .astype(int)
        )

        recovery_intervals["End_Cycle"] = (
            recovery_intervals["End_Cycle"]
            .astype(int)
        )

        # 보기 좋은 열 순서
        recovery_intervals = recovery_intervals[
            [
                "Recovery_Group",
                "Start_Cycle",
                "End_Cycle",
                "Recovery_Event_Count",
                "Start_Capacity_Ah",
                "End_Capacity_Ah",
                "Net_Interval_Recovery_Ah",
                "Net_Interval_Recovery_mAh",
                "Interval_Recovery_Percent",
                "Mean_Event_Recovery_Percent",
                "Maximum_Event_Recovery_Percent"
            ]
        ]

    else:

        recovery_intervals = pd.DataFrame(
            columns=[
                "Recovery_Group",
                "Start_Cycle",
                "End_Cycle",
                "Recovery_Event_Count",
                "Start_Capacity_Ah",
                "End_Capacity_Ah",
                "Net_Interval_Recovery_Ah",
                "Net_Interval_Recovery_mAh",
                "Interval_Recovery_Percent",
                "Mean_Event_Recovery_Percent",
                "Maximum_Event_Recovery_Percent"
            ]
        )


    # ========================================================
    # 8. 셀별 요약값 계산
    # ========================================================
    initial_capacity = df[
        "Discharge_Capacity"
    ].iloc[0]

    number_of_comparisons = max(len(df) - 1, 1)
    number_of_events = len(recovery_events)
    number_of_intervals = len(recovery_intervals)

    if number_of_events > 0:

        first_recovery_cycle = int(
            recovery_events[
                "Cycle number"
            ].iloc[0]
        )

        total_positive_recovery_mAh = (
            recovery_events[
                "Delta_Capacity_mAh"
            ].sum()
        )

        mean_event_recovery_mAh = (
            recovery_events[
                "Delta_Capacity_mAh"
            ].mean()
        )

        maximum_event_recovery_mAh = (
            recovery_events[
                "Delta_Capacity_mAh"
            ].max()
        )

        mean_event_recovery_percent = (
            recovery_events[
                "Event_Recovery_Percent"
            ].mean()
        )

        maximum_event_recovery_percent = (
            recovery_events[
                "Event_Recovery_Percent"
            ].max()
        )

        # 모든 양의 증가량 합계를 초기용량으로 나눈 값
        cumulative_recovery_vs_initial_percent = (
            total_positive_recovery_mAh
            / (initial_capacity * 1000)
            * 100
        )

    else:

        first_recovery_cycle = None
        total_positive_recovery_mAh = 0
        mean_event_recovery_mAh = 0
        maximum_event_recovery_mAh = 0
        mean_event_recovery_percent = 0
        maximum_event_recovery_percent = 0
        cumulative_recovery_vs_initial_percent = 0


    if number_of_intervals > 0:

        mean_interval_recovery_percent = (
            recovery_intervals[
                "Interval_Recovery_Percent"
            ].mean()
        )

        maximum_interval_recovery_percent = (
            recovery_intervals[
                "Interval_Recovery_Percent"
            ].max()
        )

        maximum_interval_recovery_mAh = (
            recovery_intervals[
                "Net_Interval_Recovery_mAh"
            ].max()
        )

    else:

        mean_interval_recovery_percent = 0
        maximum_interval_recovery_percent = 0
        maximum_interval_recovery_mAh = 0


    summary_list.append(
        {
            "Cell": sheet_name,
            "Initial_Capacity_Ah": initial_capacity,
            "Total_Cycles": len(df),
            "Recovery_Event_Count": number_of_events,
            "Recovery_Interval_Count": number_of_intervals,
            "Recovery_Event_Ratio_Percent":
                number_of_events
                / number_of_comparisons
                * 100,
            "First_Recovery_Cycle":
                first_recovery_cycle,
            "Total_Positive_Recovery_mAh":
                total_positive_recovery_mAh,
            "Mean_Event_Recovery_mAh":
                mean_event_recovery_mAh,
            "Maximum_Event_Recovery_mAh":
                maximum_event_recovery_mAh,
            "Mean_Event_Recovery_Percent":
                mean_event_recovery_percent,
            "Maximum_Event_Recovery_Percent":
                maximum_event_recovery_percent,
            "Mean_Interval_Recovery_Percent":
                mean_interval_recovery_percent,
            "Maximum_Interval_Recovery_Percent":
                maximum_interval_recovery_percent,
            "Maximum_Interval_Recovery_mAh":
                maximum_interval_recovery_mAh,
            "Cumulative_Recovery_vs_Initial_Percent":
                cumulative_recovery_vs_initial_percent
        }
    )


    # 결과 저장
    all_data[sheet_name] = df
    event_data[sheet_name] = recovery_events
    interval_data[sheet_name] = recovery_intervals


    # ========================================================
    # 9. 셀별 용량 그래프와 Recovery 위치
    # ========================================================
    plt.figure(figsize=(11, 6))

    plt.plot(
        df["Cycle number"],
        df["Discharge_Capacity"],
        marker="o",
        markersize=2.5,
        linewidth=1.7,
        label="Discharge capacity"
    )

    if number_of_events > 0:

        plt.scatter(
            recovery_events["Cycle number"],
            recovery_events["Discharge_Capacity"],
            marker="^",
            s=65,
            zorder=5,
            label="Capacity increase event"
        )

        # 증가율이 가장 큰 5개 지점만 표시
        largest_events = recovery_events.nlargest(
            5,
            "Event_Recovery_Percent"
        )

        for _, row in largest_events.iterrows():

            plt.annotate(
                f'{row["Event_Recovery_Percent"]:.2f}%',
                xy=(
                    row["Cycle number"],
                    row["Discharge_Capacity"]
                ),
                xytext=(3, 8),
                textcoords="offset points",
                fontsize=8
            )

    plt.xlabel("Cycle number")
    plt.ylabel("Discharge capacity (Ah)")
    plt.title(
        f"{sheet_name}: Capacity recovery events"
    )
    plt.grid(True, alpha=0.4)
    plt.legend()
    plt.tight_layout()

    plt.savefig(
        f"{sheet_name}_capacity_recovery.png",
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()


    # ========================================================
    # 10. 셀별 Recovery percentage 그래프
    # ========================================================
    plt.figure(figsize=(11, 5))

    if number_of_events > 0:

        plt.scatter(
            recovery_events["Cycle number"],
            recovery_events["Event_Recovery_Percent"],
            s=55
        )

        plt.plot(
            recovery_events["Cycle number"],
            recovery_events["Event_Recovery_Percent"],
            linewidth=1
        )

    plt.axhline(
        y=0,
        linestyle="--"
    )

    plt.xlabel("Cycle number")
    plt.ylabel("Recovery percentage (%)")
    plt.title(
        f"{sheet_name}: Recovery percentage by cycle"
    )
    plt.grid(True, alpha=0.4)
    plt.tight_layout()

    plt.savefig(
        f"{sheet_name}_recovery_percentage.png",
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()


    print("\n[개별 Recovery event]")
    display(
        recovery_events.round(5)
    )

    print("\n[연속 Recovery 구간]")
    display(
        recovery_intervals.round(5)
    )


# ============================================================
# 11. 셀별 비교 요약표
# ============================================================
summary_df = pd.DataFrame(summary_list)

summary_df = summary_df.sort_values(
    "Maximum_Event_Recovery_Percent",
    ascending=False
).reset_index(drop=True)

print("\n" + "=" * 60)
print("4개 셀 Recovery 비교 결과")
print("=" * 60)

display(
    summary_df.round(4)
)


# ============================================================
# 12. 4개 셀 Recovery percentage 중첩 그래프
# ============================================================
plt.figure(figsize=(12, 6))

for sheet_name in sheet_names:

    recovery_events = event_data[sheet_name]

    if len(recovery_events) > 0:

        plt.scatter(
            recovery_events["Cycle number"],
            recovery_events["Event_Recovery_Percent"],
            s=45,
            label=sheet_name
        )

        plt.plot(
            recovery_events["Cycle number"],
            recovery_events["Event_Recovery_Percent"],
            linewidth=1
        )

plt.axhline(
    y=0,
    linestyle="--"
)

plt.xlabel("Cycle number")
plt.ylabel("Recovery percentage (%)")
plt.title(
    "Recovery percentage comparison of NASA cells"
)
plt.grid(True, alpha=0.4)
plt.legend()
plt.tight_layout()

plt.savefig(
    "NASA_4cells_recovery_percentage.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()


# ============================================================
# 13. 셀별 평균·최대 Recovery percentage 비교
# ============================================================
x = range(len(summary_df))
bar_width = 0.35

plt.figure(figsize=(10, 6))

plt.bar(
    [i - bar_width / 2 for i in x],
    summary_df["Mean_Event_Recovery_Percent"],
    width=bar_width,
    label="Mean event recovery"
)

plt.bar(
    [i + bar_width / 2 for i in x],
    summary_df["Maximum_Event_Recovery_Percent"],
    width=bar_width,
    label="Maximum event recovery"
)

plt.xticks(
    list(x),
    summary_df["Cell"]
)

plt.xlabel("Cell")
plt.ylabel("Recovery percentage (%)")
plt.title(
    "Mean and maximum recovery percentage by cell"
)
plt.grid(
    axis="y",
    alpha=0.4
)
plt.legend()
plt.tight_layout()

plt.savefig(
    "NASA_cell_recovery_percentage_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()


# ============================================================
# 14. Recovery 횟수 및 구간 수 비교
# ============================================================
plt.figure(figsize=(10, 6))

plt.bar(
    [i - bar_width / 2 for i in x],
    summary_df["Recovery_Event_Count"],
    width=bar_width,
    label="Recovery events"
)

plt.bar(
    [i + bar_width / 2 for i in x],
    summary_df["Recovery_Interval_Count"],
    width=bar_width,
    label="Recovery intervals"
)

plt.xticks(
    list(x),
    summary_df["Cell"]
)

plt.xlabel("Cell")
plt.ylabel("Count")
plt.title(
    "Recovery event and interval count by cell"
)
plt.grid(
    axis="y",
    alpha=0.4
)
plt.legend()
plt.tight_layout()

plt.savefig(
    "NASA_cell_recovery_count_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()


# ============================================================
# 15. 전체 결과를 Excel로 저장
# ============================================================
output_name = "NASA_capacity_recovery_full_analysis.xlsx"

with pd.ExcelWriter(output_name) as writer:

    summary_df.to_excel(
        writer,
        sheet_name="Summary",
        index=False
    )

    for sheet_name in sheet_names:

        all_data[sheet_name].to_excel(
            writer,
            sheet_name=f"{sheet_name}_All",
            index=False
        )

        event_data[sheet_name].to_excel(
            writer,
            sheet_name=f"{sheet_name}_Events",
            index=False
        )

        interval_data[sheet_name].to_excel(
            writer,
            sheet_name=f"{sheet_name}_Intervals",
            index=False
        )

print("\n저장 완료:", output_name)

files.download(output_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==================================================
# 1. 파일 이름과 분석할 시트 설정
# ==================================================
file_name = "NASA Cycle DATA.xlsx"

sheet_names = ["B05", "B06", "B07", "B18"]

# EOL 기준
EOL_CAPACITY = 1.6


# ==================================================
# 2. 그래프 준비
# ==================================================
plt.figure(figsize=(11, 7))

eol_results = []


# ==================================================
# 3. 각 시트를 하나씩 불러와 그래프에 추가
# ==================================================
for sheet_name in sheet_names:

    # 해당 시트 불러오기
    cell_data = pd.read_excel(
        file_name,
        sheet_name=sheet_name
    )

    # 열 이름 앞뒤 공백 제거
    cell_data.columns = cell_data.columns.str.strip()

    # Cycle number 순서대로 정렬
    cell_data = cell_data.sort_values(
        "Cycle number"
    ).reset_index(drop=True)

    # ----------------------------------------------
    # Capacity-cycle 그래프
    # ----------------------------------------------
    plt.plot(
        cell_data["Cycle number"],
        cell_data["Discharge_Capacity"],
        marker="o",
        markersize=2.5,
        linewidth=1.8,
        label=sheet_name
    )

    # ----------------------------------------------
    # 최초 EOL 지점 찾기
    # Capacity가 1.6 Ah 미만인 첫 번째 행
    # ----------------------------------------------
    eol_data = cell_data[
        cell_data["Discharge_Capacity"] < EOL_CAPACITY
    ]

    if len(eol_data) > 0:

        first_eol = eol_data.iloc[0]

        eol_cycle = int(first_eol["Cycle number"])
        eol_capacity = first_eol["Discharge_Capacity"]

        # EOL 지점 표시
        plt.scatter(
            eol_cycle,
            eol_capacity,
            s=90,
            edgecolors="black",
            zorder=5
        )

        # EOL cycle 글자 표시
        plt.annotate(
            f"{sheet_name}: {eol_cycle} cycle",
            xy=(eol_cycle, eol_capacity),
            xytext=(eol_cycle + 3, eol_capacity - 0.04),
            arrowprops=dict(
                arrowstyle="->"
            ),
            fontsize=9
        )

        # 결과 저장
        eol_results.append(
            {
                "Cell": sheet_name,
                "EOL_cycle": eol_cycle,
                "EOL_capacity_Ah": eol_capacity
            }
        )

    else:

        eol_results.append(
            {
                "Cell": sheet_name,
                "EOL_cycle": "EOL 미도달",
                "EOL_capacity_Ah": None
            }
        )


# ==================================================
# 4. EOL 기준선
# ==================================================
plt.axhline(
    y=EOL_CAPACITY,
    linestyle="--",
    linewidth=2,
    label="EOL 기준 = 1.6 Ah"
)


# ==================================================
# 5. 그래프 설정
# ==================================================
plt.xlabel(
    "Cycle number",
    fontsize=12
)

plt.ylabel(
    "Discharge capacity (Ah)",
    fontsize=12
)

plt.title(
    "Capacity degradation and EOL comparison of NASA cells",
    fontsize=14
)

plt.legend()
plt.grid(True, alpha=0.4)
plt.tight_layout()


# ==================================================
# 6. 그래프 저장 및 출력
# ==================================================
plt.savefig(
    "NASA_4cells_capacity_EOL.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()


# ==================================================
# 7. 셀별 EOL 결과표
# ==================================================
eol_summary = pd.DataFrame(eol_results)

eol_summary = eol_summary.sort_values(
    "EOL_cycle",
    ascending=False
).reset_index(drop=True)

print("===== 셀별 최초 EOL 도달 결과 =====")
display(eol_summary)


# ==================================================
# 8. 결과표 CSV 저장
# ==================================================
eol_summary.to_csv(
    "NASA_4cells_EOL_summary.csv",
    index=False,
    encoding="utf-8-sig"
)

print("그래프 저장 완료: NASA_4cells_capacity_EOL.png")
print("결과표 저장 완료: NASA_4cells_EOL_summary.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# 1. NASA 데이터 불러오기
# ==========================================
file_name = "battery_NASA_260722_corrected.csv"

df = pd.read_csv(file_name)

# 셀과 사이클 순서로 정렬
df = df.sort_values(
    ["cell_id", "cycle_count"]
).reset_index(drop=True)

print("전체 데이터 수:", len(df))
print("셀 목록:", df["cell_id"].unique())


# ==========================================
# 2. EOL 기준 설정
# ==========================================
EOL_CAPACITY = 1.6   # Ah
EOL_SOH = 80         # %


# ==========================================
# 3. EOL 판정 열 만들기
# ==========================================
# 두 조건 중 하나라도 만족하면 EOL로 판정
df["EOL"] = (
    (df["capacity"] < EOL_CAPACITY)
    | (df["soh"] < EOL_SOH)
)


# ==========================================
# 4. 셀별 그래프 그리기
# ==========================================
eol_summary = []

for cell_id in df["cell_id"].unique():

    # 한 셀의 데이터만 선택
    cell_data = df[
        df["cell_id"] == cell_id
    ].copy()

    # --------------------------------------
    # 최초 EOL cycle 찾기
    # --------------------------------------
    eol_data = cell_data[
        cell_data["EOL"] == True
    ]

    if len(eol_data) > 0:

        # 최초 EOL 행
        first_eol = eol_data.iloc[0]

        eol_cycle = int(first_eol["cycle_count"])
        eol_capacity = first_eol["capacity"]
        eol_soh = first_eol["soh"]

        eol_summary.append(
            {
                "cell_id": cell_id,
                "EOL_cycle": eol_cycle,
                "EOL_capacity_Ah": eol_capacity,
                "EOL_SOH_percent": eol_soh
            }
        )

    else:

        eol_cycle = None
        eol_capacity = None
        eol_soh = None

        eol_summary.append(
            {
                "cell_id": cell_id,
                "EOL_cycle": "EOL 미도달",
                "EOL_capacity_Ah": None,
                "EOL_SOH_percent": None
            }
        )


    # ======================================
    # 그래프 생성
    # ======================================
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # --------------------------------------
    # 왼쪽 Y축: Capacity
    # --------------------------------------
    capacity_line = ax1.plot(
        cell_data["cycle_count"],
        cell_data["capacity"],
        marker="o",
        markersize=3,
        label="Capacity"
    )

    ax1.set_xlabel("Cycle number")
    ax1.set_ylabel("Capacity (Ah)")

    # Capacity EOL 기준선
    capacity_limit = ax1.axhline(
        y=EOL_CAPACITY,
        linestyle="--",
        label="EOL capacity = 1.6 Ah"
    )

    ax1.grid(True)


    # --------------------------------------
    # 오른쪽 Y축: SOH
    # --------------------------------------
    ax2 = ax1.twinx()

    soh_line = ax2.plot(
        cell_data["cycle_count"],
        cell_data["soh"],
        marker="s",
        markersize=3,
        label="SOH"
    )

    ax2.set_ylabel("SOH (%)")

    # SOH EOL 기준선
    soh_limit = ax2.axhline(
        y=EOL_SOH,
        linestyle=":",
        label="EOL SOH = 80%"
    )


    # --------------------------------------
    # 최초 EOL 지점 표시
    # --------------------------------------
    if eol_cycle is not None:

        # EOL cycle 수직선
        eol_vertical = ax1.axvline(
            x=eol_cycle,
            linestyle="-.",
            label=f"First EOL cycle = {eol_cycle}"
        )

        # Capacity EOL 지점 표시
        ax1.scatter(
            eol_cycle,
            eol_capacity,
            s=100,
            zorder=5
        )

        # SOH EOL 지점 표시
        ax2.scatter(
            eol_cycle,
            eol_soh,
            s=100,
            marker="X",
            zorder=5
        )

        # EOL 정보 글자로 표시
        ax1.annotate(
            (
                f"EOL cycle: {eol_cycle}\n"
                f"Capacity: {eol_capacity:.4f} Ah\n"
                f"SOH: {eol_soh:.2f}%"
            ),
            xy=(eol_cycle, eol_capacity),
            xytext=(eol_cycle + 5, eol_capacity + 0.08),
            arrowprops=dict(arrowstyle="->")
        )


    # --------------------------------------
    # 범례 합치기
    # --------------------------------------
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()

    ax1.legend(
        lines1 + lines2,
        labels1 + labels2,
        loc="upper right"
    )

    plt.title(
        f"{cell_id}: Capacity and SOH degradation"
    )

    plt.tight_layout()

    # 셀별 이미지 저장
    image_name = f"{cell_id}_Capacity_SOH_EOL.png"

    plt.savefig(
        image_name,
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()

    print(
        f"{cell_id} 그래프 저장 완료: {image_name}"
    )


# ==========================================
# 5. 셀별 EOL 결과표
# ==========================================
eol_summary_df = pd.DataFrame(eol_summary)

print("\n===== 셀별 최초 EOL 결과 =====")
display(eol_summary_df)


# ==========================================
# 6. 결과표 저장
# ==========================================
summary_file = "NASA_cell_EOL_summary.csv"

eol_summary_df.to_csv(
    summary_file,
    index=False,
    encoding="utf-8-sig"
)

print("EOL 결과표 저장 완료:", summary_file)
```

```python
import pandas as pd
from google.colab import files

# ==================================================
# 1. 데이터 불러오기
# ==================================================
df = pd.read_csv("battery_M_260722_corrected.csv")

print("전체 셀 개수:", len(df))
print("내부저항 결측치:", df["internal_resistance"].isna().sum())


# ==================================================
# 2. 내부저항에 따른 관리 등급 함수
# ==================================================
def classify_resistance(resistance):

    # 내부저항이 결측치인 경우
    if pd.isna(resistance):
        return "측정 필요"

    # 정상 구간
    elif resistance < 35:
        return "정상"

    # 관심 구간
    elif resistance < 38:
        return "관심"

    # 경고 구간
    else:
        return "경고"


# ==================================================
# 3. 관리 등급별 권고 조치 함수
# ==================================================
def recommend_action(grade):

    if grade == "정상":
        return "일반 출하 및 정상 운용"

    elif grade == "관심":
        return "동일 조건 재측정 및 용량 확인"

    elif grade == "경고":
        return "셀 격리 후 SOH·용량 추가 검사"

    else:
        return "내부저항 재측정"


# ==================================================
# 4. 모든 셀 자동 선별
# ==================================================
df["resistance_grade"] = df["internal_resistance"].apply(
    classify_resistance
)

df["recommended_action"] = df["resistance_grade"].apply(
    recommend_action
)


# ==================================================
# 5. 내부저항이 높은 순서대로 정렬
# ==================================================
df_result = df.sort_values(
    by="internal_resistance",
    ascending=False
).reset_index(drop=True)


# ==================================================
# 6. 선별 결과 확인
# ==================================================
print("\n===== 관리 등급별 셀 개수 =====")
print(df_result["resistance_grade"].value_counts())

print("\n===== 전체 선별 결과 =====")
display(
    df_result[
        [
            "cell_id",
            "internal_resistance",
            "soh",
            "resistance_grade",
            "recommended_action"
        ]
    ]
)


# ==================================================
# 7. 관심·경고 셀만 별도로 확인
# ==================================================
attention_cells = df_result[
    df_result["resistance_grade"].isin(
        ["관심", "경고"]
    )
]

print("\n===== 추가 검사가 필요한 셀 =====")
display(
    attention_cells[
        [
            "cell_id",
            "internal_resistance",
            "soh",
            "cycle_count",
            "avg_temp",
            "resistance_grade",
            "recommended_action"
        ]
    ]
)


# ==================================================
# 8. 결과 저장
# ==================================================
output_name = "battery_M_internal_resistance_screening.csv"

df_result.to_csv(
    output_name,
    index=False,
    encoding="utf-8-sig"
)

print("\n저장 완료:", output_name)

files.download(output_name)
```

```python
import pandas as pd
import numpy as np

# =========================================================
# 1. 파일 불러오기
# =========================================================
file_name = "battery_NASA_260722.csv"

df = pd.read_csv(file_name)

cell_col = "cell_id"
cycle_col = "cycle_count"
value_col = "internal_resistance"

df = df.sort_values(
    [cell_col, cycle_col]
).reset_index(drop=True)

# 원래 측정값인지, 보정값인지 구분하는 열
df["resistance_data_type"] = np.where(
    df[value_col].isna(),
    "missing",
    "measured"
)

print("보정 전 결측치:", df[value_col].isna().sum())


# =========================================================
# 2. 동일 셀의 중간 결측치
#    → 앞뒤 실측값으로 선형보간
# =========================================================
for cell_id, group in df.groupby(cell_col):

    idx = group.index
    original = df.loc[idx, value_col].copy()

    interpolated = original.interpolate(
        method="linear",
        limit_area="inside"
    )

    filled_mask = (
        original.isna()
        & interpolated.notna()
    )

    df.loc[idx, value_col] = interpolated

    filled_idx = idx[filled_mask]

    df.loc[
        filled_idx,
        "resistance_data_type"
    ] = "within-cell interpolation"


# =========================================================
# 3. 각 셀의 초기 결측 구간
#    → 다른 셀의 정규화된 증가 기울기 중앙값 사용
# =========================================================
all_cells = df[cell_col].unique()

for target_cell in all_cells:

    target = df[df[cell_col] == target_cell]

    valid_target = target[
        target[value_col].notna()
    ]

    if valid_target.empty:
        print(
            f"{target_cell}: 실측값이 없어 보정할 수 없습니다."
        )
        continue

    # 첫 번째로 존재하는 내부저항 값
    first_valid_row = valid_target.iloc[0]

    anchor_cycle = first_valid_row[cycle_col]
    anchor_value = first_valid_row[value_col]

    # 첫 실측 사이클 이전에 남아 있는 결측치
    leading_missing = target[
        (target[cycle_col] < anchor_cycle)
        & target[value_col].isna()
    ]

    if leading_missing.empty:
        continue

    first_missing_cycle = leading_missing[cycle_col].min()

    normalized_slopes = []

    # 대상 셀을 제외한 다른 셀들로 증가율 계산
    for reference_cell in all_cells:

        if reference_cell == target_cell:
            continue

        reference = df[
            df[cell_col] == reference_cell
        ][
            [cycle_col, value_col]
        ].dropna()

        # 대상 셀의 결측 구간과 같은 초기 구간 사용
        reference_window = reference[
            (reference[cycle_col] >= first_missing_cycle)
            & (reference[cycle_col] <= anchor_cycle)
        ].copy()

        # 회귀 계산을 위한 최소 데이터 수
        if len(reference_window) < 3:
            continue

        x = reference_window[cycle_col].to_numpy(
            dtype=float
        )

        y = reference_window[value_col].to_numpy(
            dtype=float
        )

        # anchor_cycle에서의 기준값 계산
        # 해당 사이클 값이 없으면 주변 값으로 선형 추정
        if not (x.min() <= anchor_cycle <= x.max()):
            continue

        reference_anchor = np.interp(
            anchor_cycle,
            x,
            y
        )

        if reference_anchor <= 0:
            continue

        # 셀별 절대저항 차이를 제거하기 위해 정규화
        y_normalized = y / reference_anchor

        x_relative = x - anchor_cycle

        # 정규화 내부저항 변화율
        slope, intercept = np.polyfit(
            x_relative,
            y_normalized,
            1
        )

        normalized_slopes.append(slope)

    if len(normalized_slopes) == 0:
        print(
            f"{target_cell}: 참고 가능한 다른 셀의 "
            "초기 데이터가 부족합니다."
        )
        continue

    # 이상치 영향을 줄이기 위해 평균 대신 중앙값 사용
    median_slope = np.median(normalized_slopes)

    print(
        f"{target_cell}: 첫 실측 사이클={anchor_cycle}, "
        f"첫 실측값={anchor_value:.4f}, "
        f"다른 셀 기반 정규화 기울기={median_slope:.6f}/cycle"
    )

    # 첫 실측값에 연결하여 이전 사이클 값 계산
    missing_cycles = leading_missing[
        cycle_col
    ].to_numpy(dtype=float)

    estimated_values = anchor_value * (
        1
        + median_slope
        * (missing_cycles - anchor_cycle)
    )

    # 물리적으로 음수가 되지 않도록 방지
    estimated_values = np.maximum(
        estimated_values,
        0
    )

    df.loc[
        leading_missing.index,
        value_col
    ] = estimated_values

    df.loc[
        leading_missing.index,
        "resistance_data_type"
    ] = "other-cell trend extrapolation"


# =========================================================
# 4. 보정 결과 확인
# =========================================================
print("\n보정 후 결측치:", df[value_col].isna().sum())

imputed_rows = df[
    df["resistance_data_type"] != "measured"
]

display(
    imputed_rows[
        [
            cell_col,
            cycle_col,
            value_col,
            "resistance_data_type"
        ]
    ].head(60)
)


# =========================================================
# 5. CSV 저장
# =========================================================
output_name = (
    "battery_NASA_internal_resistance_"
    "trend_imputed.csv"
)

df.to_csv(
    output_name,
    index=False,
    encoding="utf-8-sig"
)

print("\n저장 완료:", output_name)
```

```python
import pandas as pd

df_M = pd.read_csv("battery_M_260722.csv")

# IQR 이상치 검사를 할 변수
M_columns = [
    "initial_capacity",
    "current_capacity",
    "soh",
    "internal_resistance",
    "avg_temp",
    "charge_time"
]

# 결과를 모을 빈 리스트
M_outlier_list = []

for column in M_columns:

    # describe() 결과 저장
    stats = df_M[column].describe()

    # 1사분위수와 3사분위수
    Q1 = stats["25%"]
    Q3 = stats["75%"]

    # IQR 계산
    IQR = Q3 - Q1

    # 이상치 하한과 상한
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    print("=" * 50)
    print("변수:", column)
    print("Q1:", Q1)
    print("Q3:", Q3)
    print("IQR:", IQR)
    print("하한:", lower)
    print("상한:", upper)

    # 하한보다 작거나 상한보다 큰 행 선택
    outliers = df_M[
        (df_M[column] < lower) |
        (df_M[column] > upper)
    ]

    print("이상치 개수:", len(outliers))

    if len(outliers) > 0:
        display(outliers[["cell_id", column]])

        # 이상치 결과를 리스트에 저장
        for index, row in outliers.iterrows():
            M_outlier_list.append({
                "variable": column,
                "cell_id": row["cell_id"],
                "value": row[column],
                "lower_bound": lower,
                "upper_bound": upper
            })
```

## Q3. [읽기] SOH와 상관이 가장 강한 변수 1·2위와 계수는?
🟠 Orange3: **Correlations**
⚠️ **이상치를 제거하기 전과 후, 순위가 달라지는지 꼭 비교하세요** (Select Rows: soh ≤ 100 뒤에 Correlations를 붙여서)

```python
# ✏️ 답변 칸
# [이상치 포함]  1위:current_capacity (계수:+0.520)   2위:internal_resistance (계수:-0.391 )
# [이상치 제거후] 1위: current_capacity (계수: +0.984 )   2위:internal_resistance (계수:-0.816)
# 달라진 점 한 줄: 계수의 절댓값이 커졌다.
```

```python
# 🐍 파이썬 확인 (선택) — 제거 전/후를 나란히
print("── 이상치 포함 ──")
print(df.drop(columns=['cell_id']).corr()['soh'].sort_values().round(2))
clean = df[df['soh'] <= 100]          # 마스킹으로 이상치 제거
print("\n── 이상치 제거 후 ──")
print(clean.drop(columns=['cell_id']).corr()['soh'].sort_values().round(2))
```

## Q4. [도메인] EOL(80%) 위험 구간(80~85%)에 있는 셀을 모두 찾으면?
🟠 Orange3: **Select Rows** (soh ≥ 80 그리고 soh < 85)
🤔 **SOH가 정확히 80.0인 셀이 있다면** — EOL "미만" 기준으로 포함? 제외? 여러분이 판단하고 이유를 적으세요.

```python
# ✏️ 답변 칸
# 위험 구간 셀 ID들: B01, B03, B17
# 80.0 정각 셀에 대한 나의 판단 (포함/제외): 제외
# 이유 한 줄: 80도 성능이 떨어지기 때문
```

```python
# 🐍 파이썬 확인 (선택)
clean[(clean['soh'] >= 80) & (clean['soh'] < 85)][['cell_id', 'soh']]
```

## 필수 재현 ① · 데이터 확인
> Orange3의 **Column Statistics**로 본 것을 코드로 다시 확인합니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 데이터의 앞부분을 눈으로 확인
# [힌트] 표의 첫 몇 줄만 보여주는 메서드
# ─────────────────────────────────────────────
import pandas as pd

# 1. 파일 불러오기 (파일명은 실제 업로드한 파일명)
df = pd.read_csv('battery_M_260722.csv')

# 2. 첫 5줄 보기 (기본값: 5줄)
df.head()
````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 각 컬럼의 타입과 결측(Non-Null) 개수 확인
# [힌트] "출석부 훑기" — 교재 2-3장의 세 버튼 중 두 번째
# [결과] 어떤 컬럼에 결측이 몇 개인지 알 수 있어야 함
# ─────────────────────────────────────────────


# ✏️ 확인한 내용 (발표 3·4장 재료):
#    결측 컬럼: current_capacity     개수: 3개
import pandas as pd

# 파일 불러오기 (파일명/경로)
df = pd.read_csv('battery_M_260722.csv')

# 각 컬럼의 타입, Non-Null 개수, 전체 데이터 정보 출력
df.info()
````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 수치 컬럼의 요약 통계로 이상한 값 찾기
# [힌트] 세 버튼 중 세 번째. min/max를 유심히 볼 것
# [결과] 물리적으로 불가능한 값이 눈에 띄어야 함
# ─────────────────────────────────────────────


# ✏️ 이상치: soh 컬럼      121 값    왜 불가능한가: soh값은 100초과 불가
import pandas as pd
import numpy as np

# 1. 수치형 컬럼만 선택
numeric_df = df.select_dtypes(include=[np.number])

# 2. 기본 기술통계량 구하기 (T를 붙여 행/열 전환)
summary = numeric_df.describe().T

# 3. IQR 및 상/하한 경계선 계산
iqr = summary['75%'] - summary['25%']
lower_bound = summary['25%'] - (1.5 * iqr)
upper_bound = summary['75%'] + (1.5 * iqr)

# 4. 이상치(Outlier) 개수 구하는 함수
def get_outlier_count(col):
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr_val = q3 - q1
    l_bound = q1 - 1.5 * iqr_val
    u_bound = q3 + 1.5 * iqr_val
    return ((col < l_bound) | (col > u_bound)).sum()

# 5. 요약표에 이상치 개수 및 0(Zero) 개수 추가
summary['Zero_Count'] = numeric_df.apply(lambda x: (x == '').sum())
summary['Outlier_Count'] = numeric_df.apply(get_outlier_count)

# 6. 보기 편하게 주요 컬럼 순서 정리
result_df = summary[['count', 'mean', '50%', 'min', 'max', 'std', 'Zero_Count', 'Outlier_Count']]
result_df.columns = ['전체개수', '평균값', '중앙값(50%)', '최소값', '최대값', '표준편차', '0의개수', '이상치개수(IQR)']

# 결과 출력
result_df
````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 이상치를 제외한 데이터를 m_clean 이라는 이름으로 만들기
# [힌트] 조건을 걸어 골라내기 = 마스킹 (교재 1-5, 2-5장)
#        SOH는 100%를 넘을 수 없다
# [결과] print로 제거 전/후 행 수를 비교해 보세요
# ─────────────────────────────────────────────
import pandas as pd
import numpy as np

# 1. 수치형 컬럼 목록 추출
numeric_cols = df.select_dtypes(include=[np.number]).columns

# 2. 모든 수치형 컬럼에 대해 IQR 이상치 여부(Boolean Matrix) 판단
is_outlier = pd.DataFrame(False, index=df.index, columns=numeric_cols)

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    l_bound = Q1 - 1.5 * IQR
    u_bound = Q3 + 1.5 * IQR

    # 해당 컬럼의 범위를 벗어난 지점 체크
    is_outlier[col] = (df[col] < l_bound) | (df[col] > u_bound)

# 3. 어느 한 컬럼에서라도 이상치가 존재하지 '않는' 행만 선택해 m_clean 생성
m_clean = df[~is_outlier.any(axis=1)].copy()

# 4. 제거 전/후 행 수 비교 및 출력
before_rows = len(df)
after_rows = len(m_clean)
removed_rows = before_rows - after_rows

print("=== 모든 수치형 컬럼 종합 이상치 제거 결과 ===")
print(f"제거 전 행 수 : {before_rows:,} 개")
print(f"제거 후 행 수 : {after_rows:,} 개 (m_clean)")
print(f"제거된 행 수  : {removed_rows:,} 개 ({removed_rows / before_rows * 100:.2f}%)")

````

## 3단계 · 열화 곡선 그리기 ★ 오늘의 핵심 그래프
> Orange3에서는 산점도로 봤지만, 파이썬에서는 **진짜 선 그래프**를 그릴 수 있습니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 셀마다 선을 하나씩 그려서, 사이클에 따른 SOH 변화 표시
#
# [설계 힌트]
#   1) 셀별로 데이터를 나눠야 한다 → groupby로 묶으면 (이름, 데이터) 쌍이 나옵니다
#   2) 각 셀마다 선을 하나씩 → plt.plot(x축값, y축값, label=셀이름)
#   3) EOL 기준선 80%를 가로선으로 → plt.axhline(80, color='red', linestyle='--')
#   4) 범례·축 이름 → plt.legend(), plt.xlabel(), plt.ylabel()
#   5) 마지막에 plt.show()
#
# [결과] 왼쪽 위에서 오른쪽 아래로 흘러내리는 곡선 4개 + 빨간 기준선
# ─────────────────────────────────────────────


# ✏️ 이 그림에서 알게 된 것 한 문장 (발표 8장):
import pandas as pd
import matplotlib.pyplot as plt

# 1. Colab 경로에 있는 CSV 파일 불러오기
file_path = '/content/battery_NASA_260722.csv'
df = pd.read_csv(file_path)

# [주의] 에러 방지를 위해 실제 데이터의 컬럼명을 확인하고 아래 변수에 입력해 주세요.
# 만약 파일의 컬럼명이 소문자 'cell', 'cycle', 'soh'라면 아래 문자열을 그에 맞게 수정해야 합니다.
cell_col = 'internal_resistance'
cycle_col = 'cycle_count'
soh_col = 'soh'

plt.figure(figsize=(10, 6))

# 1) 셀별로 데이터를 나누기 (groupby)
# 2) 각 셀마다 선을 하나씩 그리기 (plot)
for cell_name, cell_data in df.groupby(cell_col):
    plt.plot(cell_data[cycle_col], cell_data[soh_col], label=cell_name)

# 3) EOL 기준선 80%를 가로선으로 그리기
plt.axhline(80, color='red', linestyle='--', label='EOL 80%')

# 4) 범례 및 축 이름 설정
plt.legend()
plt.xlabel('Cycle')
plt.ylabel('SOH (%)')
plt.title('Battery SOH Degradation over Cycles')

# 5) 그래프 출력
plt.show()
````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 각 셀이 몇 번째 사이클에 EOL(SOH 80% 미만)에 도달했는지
# [힌트] EOL 여부를 나타내는 컬럼이 이미 있습니다 (is_eol)
#        그 값이 1인 행만 고른 뒤, 셀별로 묶어서 가장 작은 사이클을 찾으면 됩니다
# [결과] 셀 4개의 EOL 도달 사이클
# ─────────────────────────────────────────────


# ✏️ 가장 빨리 죽은 셀:          가장 오래 버틴 셀:          차이:      사이클
````

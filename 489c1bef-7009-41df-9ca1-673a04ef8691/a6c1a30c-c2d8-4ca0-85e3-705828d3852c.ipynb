## 필수 재현 ② · 최종 판정
> Orange3의 **Select Rows**로 한 선별을 코드로 재현합니다. **두 결과가 같아야 정상입니다.**

````python
# ─────────────────────────────────────────────
# [해야 할 일] 내가 Orange3에서 정한 기준을 조건으로 옮겨 최종 셀 선별
# [힌트] 조건이 2개 이상이면 각각 괄호로 감싸고 & 로 연결
#        df[(조건1) & (조건2)]  ← 괄호 빠뜨리면 에러 (오늘의 1등 에러)
# [결과] 선별된 셀 수를 출력하고, cell_id와 soh를 확인
# ─────────────────────────────────────────────
import pandas as pd

# 1. CSV 데이터 로드
df = pd.read_csv('battery_M_260722.csv')

# 2. current_capacity 비어있는 칸 수식 적용하여 채우기
# 수식 변형: current_capacity = (SOH * initial_capacity) / 100
missing_cap_mask = df['current_capacity'].isnull()
df.loc[missing_cap_mask, 'current_capacity'] = (df.loc[missing_cap_mask, 'soh'] * df.loc[missing_cap_mask, 'initial_capacity']) / 100

# 3. SOH 이상치(100 초과) 수정하기
# 수식: SOH = (current_capacity / initial_capacity) * 100
invalid_soh_mask = df['soh'] > 100
df.loc[invalid_soh_mask, 'soh'] = (df.loc[invalid_soh_mask, 'current_capacity'] / df.loc[invalid_soh_mask, 'initial_capacity']) * 100

# 4. SOH 85 초과 셀의 cell_id와 soh 추출
result = df[df['soh'] > 85][['cell_id', 'soh']]

# 보기 쉽게 SOH 소수점 둘째 자리 반올림
result['soh'] = result['soh'].round(2)

# 결과 출력
print(result)

````

### "그 기준, 흔들리면 결과가 얼마나 달라지나요?"

기준값 하나로 판정하고 끝내면 평범합니다. **기준이 얼마나 예민한지 수치로** 보여주세요.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 내 기준값을 -3 / 그대로 / +3 으로 바꿔가며
#              각각 몇 개가 통과하는지 표처럼 출력
#
# [생각해볼 것] 같은 코드를 값만 바꿔 3번 쓰는 것보다,
#               값 목록을 만들어 반복시키는 게 낫지 않을까?
#               (어제 배운 for문 — 리스트를 하나씩 꺼내 쓰기)
#
# [결과 예시]
#   soh >= 82: 통과 26개 / 탈락 3개
#   soh >= 85: 통과 24개 / 탈락 5개
#   soh >= 88: 통과 22개 / 탈락 7개
# ─────────────────────────────────────────────
import pandas as pd

# 1. 데이터 로드 및 전처리 (이전과 동일하게 보정)
df = pd.read_csv('battery_M_260722.csv')

# 결측치 수식 보정
missing_cap_mask = df['current_capacity'].isnull()
df.loc[missing_cap_mask, 'current_capacity'] = (df.loc[missing_cap_mask, 'soh'] * df.loc[missing_cap_mask, 'initial_capacity']) / 100

# 이상치 SOH 수식 보정
invalid_soh_mask = df['soh'] > 100
df.loc[invalid_soh_mask, 'soh'] = (df.loc[invalid_soh_mask, 'current_capacity'] / df.loc[invalid_soh_mask, 'initial_capacity']) * 100

# 2. 기준값 목록 생성 (기준 88 기준: 85, 88, 91 + 추가 요구사항 90)
thresholds = [85, 88, 90]  # 또는 기준값에 맞춰 [85, 88, 91]
total_cells = len(df)          # 전체 셀 수 (30개)

# 3. for문으로 조건별 통과/탈락 셀 수 계산 및 출력
for threshold in thresholds:
    passed = (df['soh'] > threshold).sum()
    failed = total_cells - passed
    print(f"soh > {threshold}: 통과 {passed}개 / 탈락 {failed}개")

# ✏️ 발표 7장 문장: "기준을 3%p 낮추면 __개가 더 통과합니다.
#                    그만큼 예민한 경계이므로 __를 근거로 __로 정했습니다"
````

# 관문 ② · 결측의 영향 ★변별
### "빈칸을 채운 게 답을 바꾸지 않았나요?"

결측을 평균으로 채우고 넘어가면 놓치는 게 있습니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일 1] 결측이 있던 셀들이 어떤 값을 가졌는지 확인
# [힌트] 결측 여부를 판단하는 메서드가 있습니다 (NaN인가?)
#        그 결과를 마스킹 조건으로 쓰면 결측 행만 골라집니다
# [보고 싶은 것] 그 셀들의 cell_id, soh, internal_resistance
# ─────────────────────────────────────────────
import pandas as pd

# 1. 원본 데이터 로드
df = pd.read_csv('battery_M_260722.csv')

# 2. 문제 셀 조건 설정 (마스킹)
# 조건 A: 비어있는 칸(결측치)이 있는 행 -> df.isna().any(axis=1) 또는 df.isnull().any(axis=1)
# 조건 B: SOH가 100%를 초과하는 이상치 행 -> df['soh'] > 100
mask_null = df.isna().any(axis=1)
mask_invalid_soh = df['soh'] > 100

# 3. 두 조건을 만족하는(또는 관계) 문제 셀 추출
problem_cells = df[mask_null | mask_invalid_soh][['cell_id', 'soh', 'internal_resistance']]

# 결과 출력
print(problem_cells)

# ✏️ 결측 셀들의 SOH는 합격 경계 근처인가, 확실한 불합격인가?
# M12의 경우 soh를 새로 구해야함, 나머지의 경우 soh 85 초과 기준에서는 3가지 모두 만족, 90초과 기준에서는 1가지만 만족
````

````python
# ─────────────────────────────────────────────
# [해야 할 일 2] 결측을 두 가지 방식으로 처리하고 최종 판정을 비교
#   ⓐ 평균으로 채우기  → 내 기준 적용 → 몇 개 통과?
#   ⓑ 해당 행 제외하기 → 내 기준 적용 → 몇 개 통과?
#
# [힌트] 빈칸 채우기 / 빈칸 있는 행 버리기에 해당하는 메서드가 각각 있습니다
#        (교재 4-9 로제타석의 '결측 채우기' 행 참고)
# [결과] 두 방식의 최종 셀 수를 나란히 출력
# ─────────────────────────────────────────────
import pandas as pd

# 1. 원본 데이터 불러오기
df = pd.read_csv('battery_M_260722.csv')

# --- ⓐ 결측치 및 이상치(SOH > 100)를 평균으로 대체하기 ---
df_a = df.copy()

# 1) current_capacity 결측치를 평균값으로 대체
df_a['current_capacity'] = df_a['current_capacity'].fillna(df_a['current_capacity'].mean())

# 2) SOH가 100 이하인 정상 데이터들의 평균 계산 (약 92.20%)
valid_soh_mean = df_a.loc[df_a['soh'] <= 100, 'soh'].mean()

# 3) SOH가 100을 초과하는 이상치(M12)를 SOH 평균값으로 대체
df_a.loc[df_a['soh'] > 100, 'soh'] = valid_soh_mean

# SOH > 85 초과 통과 개수 산출
pass_a = (df_a['soh'] > 85).sum()


# --- ⓑ 결측치 및 이상치(SOH > 100) 행 완전히 제외하기 ---
df_b = df.dropna().copy()
df_b = df_b[df_b['soh'] <= 100]  # SOH 100 초과 이상치 행도 삭제

# SOH > 85 초과 통과 개수 산출
pass_b = (df_b['soh'] > 85).sum()


# --- 결과 출력 ---
print(f"ⓐ 평균 대체 방식 : 통과 {pass_a}개 (전체 {len(df_a)}개 중)")
print(f"ⓑ 행 삭제 방식   : 통과 {pass_b}개 (전체 {len(df_b)}개 중)")

# ✏️ 판정이 달라졌는가? 나는 어느 쪽을 택했고 왜인가:
# 단순히 값이 누락되었거나 측정 오류가 있다고 해서 셀을 바로 버리는 것은 어쩌면 경제적인 문제와도 연결될 수 있기 때문에 평균으로 대체해서
# 더 많은 데이터를 이용하는 것이 경제적이기는 하나 정확하게 측정된 데이터는 아니므로 반드시 측정 및 재검토가 이루어져야한다.
````

### "상관 1위가 정말 원인인가요?"

이 데이터에는 **의도적으로 심어둔 함정**이 있습니다. 순위표만 보면 걸립니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 이상치를 "제거하기 전(m)"과 "제거한 후(m_clean)"의
#              SOH 상관 순위를 각각 구해서 비교
#
# [힌트] 상관은 corr() — 단, cell_id는 글자라 계산에서 빼야 합니다
#        특정 열을 빼는 방법: drop(columns=[...])
#        SOH 기준 순위만 보려면 결과에서 'soh' 열을 뽑고 정렬
# [결과] 두 순위표를 나란히 출력해 어떤 변수의 순위가 크게 움직였는지 확인
# ─────────────────────────────────────────────
import pandas as pd

# 1. 원본 데이터 로드 (이상치 제거 전 m)
m = pd.read_csv('battery_M_260722.csv')

# 2. 이상치 및 결측치 제거 데이터 생성 (이상치 제거 후 m_clean)
# - 결측치(dropna) 및 SOH 100% 초과 이상치 제거
m_clean = m.dropna().copy()
m_clean = m_clean[m_clean['soh'] <= 100]

# 3. cell_id(문자열) 제외 후 SOH 기준 상관계수 계산 및 정렬
corr_before = m.drop(columns=['cell_id']).corr()['soh'].sort_values(ascending=False)
corr_after = m_clean.drop(columns=['cell_id']).corr()['soh'].sort_values(ascending=False)

# 4. 두 결과를 하나의 데이터프레임으로 나란히 결합
result = pd.DataFrame({
    '제거 전 (m)': corr_before,
    '제거 후 (m_clean)': corr_after
})

print("=== SOH와의 상관관계 비교표 ===")
print(result)

````

## 3단계 · 열화 곡선 그리기 ★ 오늘의 핵심 그래프
> Orange3에서는 산점도로 봤지만, 파이썬에서는 **진짜 선 그래프**를 그릴 수 있습니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 셀마다 선을 하나씩 그려서, 사이클에 따른 SOH 변화 표시
#
# [설계 힌트]
#   1) 셀별로 데이터를 나눠야 한다 → groupby로 묶으면 (이름, 데이터) 쌍이 나옵니다
#   2) 각 셀마다 선을 하나씩 → plt.plot(x축값, y축값, label=셀이름)
#   3) EOL 기준선 80%를 가로선으로 → plt.axhline(80, color='red', linestyle='--')
#   4) 범례·축 이름 → plt.legend(), plt.xlabel(), plt.ylabel()
#   5) 마지막에 plt.show()
#
# [결과] 왼쪽 위에서 오른쪽 아래로 흘러내리는 곡선 4개 + 빨간 기준선
# ─────────────────────────────────────────────


# ✏️ 이 그림에서 알게 된 것 한 문장 (발표 8장):
````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 각 셀이 몇 번째 사이클에 EOL(SOH 80% 미만)에 도달했는지
# [힌트] EOL 여부를 나타내는 컬럼이 이미 있습니다 (is_eol)
#        그 값이 1인 행만 고른 뒤, 셀별로 묶어서 가장 작은 사이클을 찾으면 됩니다
# [결과] 셀 4개의 EOL 도달 사이클
# ─────────────────────────────────────────────
import pandas as pd

# 1. 파일 데이터 로드
df = pd.read_csv('battery_NASA_260722.csv')

# 2. EOL(SOH 80% 미만) 상태인 행만 필터링 (is_eol == 1)
eol_df = df[df['is_eol'] == 1]

# 3. 셀(cell_id)별로 그룹화하여 가장 작은(최초) 사이클 번호 찾기
eol_cycles = eol_df.groupby('cell_id')['cycle_count'].min().reset_index()

# 4. 결과 출력
print("=== 셀 4개의 EOL 도달 사이클 ===")
print(eol_cycles)

# 5. 가장 빨리 죽은 셀, 가장 오래 버틴 셀 및 차이 계산
min_row = eol_cycles.loc[eol_cycles['cycle_count'].idxmin()]
max_row = eol_cycles.loc[eol_cycles['cycle_count'].idxmax()]
diff = max_row['cycle_count'] - min_row['cycle_count']

print(f"\n가장 빨리 죽은 셀: {min_row['cell_id']} ({min_row['cycle_count']} 사이클)")
print(f"가장 오래 버틴 셀: {max_row['cell_id']} ({max_row['cycle_count']} 사이클)")
print(f"차이: {diff} 사이클")

# ✏️ 가장 빨리 죽은 셀:          가장 오래 버틴 셀:          차이:      사이클
````

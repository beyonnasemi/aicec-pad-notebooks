## 필수 재현 ② · 최종 판정

# 관문 ① · 기준 민감도 ★변별
### "그 기준, 흔들리면 결과가 얼마나 달라지나요?"

기준값 하나로 판정하고 끝내면 평범합니다. **기준이 얼마나 예민한지 수치로** 보여주세요.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 내 기준값을 -3 / 그대로 / +3 으로 바꿔가며
#              각각 몇 개가 통과하는지 표처럼 출력
#
# [생각해볼 것] 같은 코드를 값만 바꿔 3번 쓰는 것보다,
#               값 목록을 만들어 반복시키는 게 낫지 않을까?
#               (어제 배운 for문 — 리스트를 하나씩 꺼내 쓰기)
#
# [결과 예시]
#   soh >= 82: 통과 26개 / 탈락 3개
#   soh >= 85: 통과 24개 / 탈락 5개
#   soh >= 88: 통과 22개 / 탈락 7개
# ─────────────────────────────────────────────


# ✏️ 발표 7장 문장: "기준을 3%p 낮추면 2개가 더 통과합니다.
#                    그만큼 예민한 경계이므로 85%를 비교용 운영 기준으로 정했습니다."

for cutoff in [SOH_CUTOFF - 3, SOH_CUTOFF, SOH_CUTOFF + 3]:
    passed = (m_clean["soh"] >= cutoff).sum()
    failed = len(m_clean) - passed
    print(f"soh >= {cutoff}: 통과 {passed}개 / 탈락 {failed}개")

# 작성: 기준을 3%p 낮추면 2개가 더 통과합니다.
#       판정이 기준값에 민감하므로 85%를 비교용 운영 기준으로 정했습니다.
````

````python
# ─────────────────────────────────────────────
# [해야 할 일 · 심화] 내 기준 경계에 아슬아슬하게 걸린 셀 찾기
# [힌트] 기준값 ±2 범위에 있는 셀만 골라내면 됩니다 (조건 2개)
# [왜?] "이 셀은 왜 탈락시켰나요?" 질문에 사례로 답할 수 있습니다
# ─────────────────────────────────────────────


borderline = m_clean[
    (m_clean["soh"] >= SOH_CUTOFF - 2) &
    (m_clean["soh"] <= SOH_CUTOFF + 2)
][["cell_id", "soh", "internal_resistance"]].sort_values("soh")

display(borderline)
````

# 관문 ② · 결측의 영향 ★변별
### "빈칸을 채운 게 답을 바꾸지 않았나요?"

결측을 평균으로 채우고 넘어가면 놓치는 게 있습니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일 1] 결측이 있던 셀들이 어떤 값을 가졌는지 확인
# [힌트] 결측 여부를 판단하는 메서드가 있습니다 (NaN인가?)
#        그 결과를 마스킹 조건으로 쓰면 결측 행만 골라집니다
# [보고 싶은 것] 그 셀들의 cell_id, soh, internal_resistance
# ─────────────────────────────────────────────


# ✏️ 결측 셀들의 SOH는 합격 경계 근처인가, 확실한 불합격인가?

missing_cells = m_clean[m_clean["current_capacity"].isna()][
    ["cell_id", "soh", "internal_resistance"]
]
display(missing_cells)

# 작성: 결측 셀들은 모두 85% 기준을 통과하며, M18은 경계에 비교적 가까움
````

````python
# ─────────────────────────────────────────────
# [해야 할 일 2] 결측을 두 가지 방식으로 처리하고 최종 판정을 비교
#   ⓐ 평균으로 채우기  → 내 기준 적용 → 몇 개 통과?
#   ⓑ 해당 행 제외하기 → 내 기준 적용 → 몇 개 통과?
#
# [힌트] 빈칸 채우기 / 빈칸 있는 행 버리기에 해당하는 메서드가 각각 있습니다
#        (교재 4-9 로제타석의 '결측 채우기' 행 참고)
# [결과] 두 방식의 최종 셀 수를 나란히 출력
# ─────────────────────────────────────────────


# ✏️ 판정이 달라졌는가? 나는 어느 쪽을 택했고 왜인가:
# 작성: SOH 자체는 바뀌지 않지만 행 제외 시 표본이 3개 줄어듦.
#       따라서 행은 유지하고 current_capacity는 독립 상관변수 해석에서 제외함.

m_mean = m_clean.copy()
m_mean["current_capacity"] = m_mean["current_capacity"].fillna(
    m_mean["current_capacity"].mean()
)
passed_mean = len(m_mean[m_mean["soh"] >= SOH_CUTOFF])

m_drop = m_clean.dropna(subset=["current_capacity"]).copy()
passed_drop = len(m_drop[m_drop["soh"] >= SOH_CUTOFF])

print("평균 대치 후 통과:", passed_mean, "개")
print("결측 행 제외 후 통과:", passed_drop, "개")


````

# 관문 ③ · 함정 회피 ★변별
### "상관 1위가 정말 원인인가요?"

이 데이터에는 **의도적으로 심어둔 함정**이 있습니다. 순위표만 보면 걸립니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 이상치를 "제거하기 전(m)"과 "제거한 후(m_clean)"의
#              SOH 상관 순위를 각각 구해서 비교
#
# [힌트] 상관은 corr() — 단, cell_id는 글자라 계산에서 빼야 합니다
#        특정 열을 빼는 방법: drop(columns=[...])
#        SOH 기준 순위만 보려면 결과에서 'soh' 열을 뽑고 정렬
# [결과] 두 순위표를 나란히 출력해 어떤 변수의 순위가 크게 움직였는지 확인
# ─────────────────────────────────────────────



before_corr = (
    m.drop(columns=["cell_id"])
     .corr(numeric_only=True)["soh"]
     .drop("soh")
)
before_corr = before_corr.reindex(
    before_corr.abs().sort_values(ascending=False).index
)

after_corr = (
    m_clean.drop(columns=["cell_id"])
           .corr(numeric_only=True)["soh"]
           .drop("soh")
)
after_corr = after_corr.reindex(
    after_corr.abs().sort_values(ascending=False).index
)

display(pd.concat(
    [before_corr.rename("제거 전"), after_corr.rename("M12 제거 후")],
    axis=1
))
````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 의심스러운 변수를 산점도로 다시 확인
# [힌트] 산점도는 plot.scatter(x=..., y=...)
# [따져볼 것] 이 변수가 배터리를 늙게 할 "물리적 이유"가 있는가?
#             (예: 생산 배치 번호가 커지면 배터리가 빨리 늙는다? 말이 되는가?)
# ─────────────────────────────────────────────


# ✏️ 상관 1위 변수:              물리적으로 타당한가?
# ✏️ 함정으로 의심되는 변수:              배제 이유:
# 작성: 상관 1위 변수 current_capacity — SOH 계산식에 직접 포함되어 독립 원인으로 보기 어려움
# 작성: 물리적으로 타당한 변수 internal_resistance — 저항 증가와 SOH 감소가 강한 음의 상관

m_clean.plot.scatter(x="internal_resistance", y="soh", figsize=(7, 5))
plt.title("Internal Resistance vs SOH")
plt.xlabel("Internal Resistance")
plt.ylabel("SOH (%)")
plt.grid(alpha=0.3)
plt.show()

print("Pearson:", round(m_clean["internal_resistance"].corr(m_clean["soh"]), 3))
print("Spearman:", round(m_clean["internal_resistance"].corr(
    m_clean["soh"], method="spearman"
), 3))


````

## 2단계 · 데이터 확인

````python
# ─────────────────────────────────────────────
# [해야 할 일] NASA 데이터의 구조 파악 — 앞부분 + 결측 확인
# [주목] 결측이 있다면 어느 컬럼에, 몇 개나?
# ─────────────────────────────────────────────



display(nasa.head())
nasa.info()

print("\n컬럼별 결측 개수")
print(nasa.isna().sum())
````

````python
# ─────────────────────────────────────────────
# [해야 할 일 1] 셀마다 사이클 수가 다른지 확인
# [힌트] 셀별로 묶어서(groupby) 각 그룹의 최대 사이클을 보면 됩니다
# [왜?] 셀마다 관측 길이가 다르면 비교할 때 주의해야 합니다 (발표 '한계' 재료)
# ─────────────────────────────────────────────


# ─────────────────────────────────────────────
# [해야 할 일 2 · 관문② 심화] 결측이 어디에 있는지 확인 ★
# [힌트] 결측인 행만 골라낸 뒤, 셀별·사이클별로 어디에 몰려 있는지 보세요
#        (groupby로 묶어 min/max 사이클을 보면 구간이 드러납니다)
# [따져볼 것] 결측이 무작위인가, 특정 구간에 몰려 있는가?
#             몰려 있다면 평균으로 채우는 게 왜 위험한가?
# ─────────────────────────────────────────────


# ✏️ 결측 구간:                     평균 대체가 위험한 이유:
# 작성: B0005, B0006, B0007의 1~19사이클에 각각 19개씩 결측
# 작성: 초기 구간의 체계적 결측이라 평균 대체 시 초기 저항 추세가 왜곡될 수 있음

max_cycles = nasa.groupby("cell_id")["cycle_count"].max()
print("셀별 최대 사이클")
display(max_cycles)

missing_ir = nasa[nasa["internal_resistance"].isna()]
missing_range = missing_ir.groupby("cell_id")["cycle_count"].agg(
    ["count", "min", "max"]
)
print("내부저항 결측 구간")
display(missing_range)


````

## 3단계 · 열화 곡선 그리기 ★ 오늘의 핵심 그래프
> Orange3에서는 산점도로 봤지만, 파이썬에서는 **진짜 선 그래프**를 그릴 수 있습니다.

````python
# ─────────────────────────────────────────────
# [해야 할 일] 셀마다 선을 하나씩 그려서, 사이클에 따른 SOH 변화 표시
#
# [설계 힌트]
#   1) 셀별로 데이터를 나눠야 한다 → groupby로 묶으면 (이름, 데이터) 쌍이 나옵니다
#   2) 각 셀마다 선을 하나씩 → plt.plot(x축값, y축값, label=셀이름)
#   3) EOL 기준선 80%를 가로선으로 → plt.axhline(80, color='red', linestyle='--')
#   4) 범례·축 이름 → plt.legend(), plt.xlabel(), plt.ylabel()
#   5) 마지막에 plt.show()
#
# [결과] 왼쪽 위에서 오른쪽 아래로 흘러내리는 곡선 4개 + 빨간 기준선
# ─────────────────────────────────────────────


# ✏️ 이 그림에서 알게 된 것 한 문장 (발표 8장):
# 작성: 네 셀 모두 사이클 증가에 따라 SOH가 감소하지만 EOL 도달 시점은 셀마다 다르다.

plt.figure(figsize=(10, 6))

for cell_id, group in nasa.groupby("cell_id"):
    group = group.sort_values("cycle_count")
    plt.plot(group["cycle_count"], group["soh"], label=cell_id)

plt.axhline(80, color="red", linestyle="--", label="EOL 80%")
plt.legend()
plt.xlabel("Cycle Count")
plt.ylabel("SOH (%)")
plt.title("NASA Battery SOH Degradation")
plt.grid(alpha=0.3)
plt.show()


````

````python
# ─────────────────────────────────────────────
# [해야 할 일] 각 셀이 몇 번째 사이클에 EOL(SOH 80% 미만)에 도달했는지
# [힌트] EOL 여부를 나타내는 컬럼이 이미 있습니다 (is_eol)
#        그 값이 1인 행만 고른 뒤, 셀별로 묶어서 가장 작은 사이클을 찾으면 됩니다
# [결과] 셀 4개의 EOL 도달 사이클
# ─────────────────────────────────────────────


# ✏️ 가장 빨리 죽은 셀:          가장 오래 버틴 셀:          차이:      사이클
# 작성: 가장 빨리 B0018 / 가장 오래 B0007 / 차이 41사이클

eol_cycle = (
    nasa[nasa["is_eol"] == 1]
    .groupby("cell_id")["cycle_count"]
    .min()
)
display(eol_cycle)

fastest_cell = eol_cycle.idxmin()
slowest_cell = eol_cycle.idxmax()
cycle_gap = int(eol_cycle.max() - eol_cycle.min())

print("가장 빨리 죽은 셀:", fastest_cell)
print("가장 오래 버틴 셀:", slowest_cell)
print("차이:", cycle_gap, "사이클")

````

### "PART 1의 발견이 PART 2에서도 성립하나요?"

두 데이터를 따로 분석하고 끝내면 절반입니다. **연결**하세요.

````python
# ─────────────────────────────────────────────
# [해야 할 일] PART 1에서 찾은 열화 신호(변수)가
#              NASA 데이터에서도 같은 방향인지 상관으로 확인
#
# [힌트] 두 데이터 각각에서 해당 변수와 soh의 상관계수를 구해 나란히 출력
# [따져볼 것] 방향(+/-)은 같은가? 강도는? 왜 차이가 날까?
# ─────────────────────────────────────────────


# ✏️ M데이터 상관:          NASA 상관:
# ✏️ 나의 해석 (발표 9장):
# 작성: 두 데이터 모두 음의 방향이다.
#       M은 셀 간 스냅샷, NASA는 반복 시계열이므로 상관 강도에는 차이가 있다.

m_corr = m_clean["internal_resistance"].corr(m_clean["soh"])

nasa_valid = nasa.dropna(subset=["internal_resistance"])
nasa_corr = nasa_valid["internal_resistance"].corr(nasa_valid["soh"])

print("M데이터 상관:", round(m_corr, 3))
print("NASA 상관:", round(nasa_corr, 3))
````

````python
# ─────────────────────────────────────────────
# [해야 할 일 · 심화] 셀별로 나눠서 본 상관 vs 전체를 한 덩어리로 본 상관 ★
# [힌트] groupby로 셀을 나눈 뒤, 각 그룹에서 저항-SOH 상관을 구해 비교
# [따져볼 것] 셀마다 값이 크게 다른가? 전체 상관과 차이가 나는가?
#             다르다면 "평균의 함정"일 수 있습니다
# ─────────────────────────────────────────────


# ─────────────────────────────────────────────
# [해야 할 일 · 심화] 셀별 열화 속도(기울기) 비교
# [힌트] (첫 SOH - 마지막 SOH) / 사이클 수, 또는 numpy polyfit으로 직선 맞추기
# ─────────────────────────────────────────────



cell_corr = {}
for cell_id, group in nasa_valid.groupby("cell_id"):
    cell_corr[cell_id] = group["internal_resistance"].corr(group["soh"])

cell_corr = pd.Series(cell_corr, name="저항-SOH 상관")
display(cell_corr.to_frame())

slopes = {}
for cell_id, group in nasa.groupby("cell_id"):
    group = group.sort_values("cycle_count")
    slopes[cell_id] = np.polyfit(
        group["cycle_count"], group["soh"], 1
    )[0]

slope_table = pd.Series(slopes, name="사이클당 SOH 기울기")
display(slope_table.to_frame())

print("NASA 전체 상관:", round(nasa_corr, 3))
````
